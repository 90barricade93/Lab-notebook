## Start

```python
# import libraries
import pandas as pd
import plotly.express as px
```

## Task Load dataset into Python

```python
# load datasets
WijnRood = pd.read_csv('WDS\winequality-red.csv', delimiter=';')
WijnWit = pd.read_csv('WDS\winequality-white.csv', delimiter=';')
```

## Task Print out the dataset head and data types

```python
# merge datasets into one dataframe 
Wtotaal = pd.concat([WijnRood, WijnWit])
Wtotaal.head()
```

```python
# obtain datatypes
print("Data types:\n\n" + str(Wtotaal.dtypes))	
```

## Task Obtain value counts for each feature in the dataset

```python
# obtain value counts per feature
print("Alcohol:\n\n", Wtotaal['alcohol'].value_counts(), "\n\n",
      "Sulphates:\n\n", Wtotaal['sulphates'].value_counts(), "\n\n",
      "Density:\n\n", Wtotaal['density'].value_counts(), "\n\n", 
      "pH:\n\n", Wtotaal['pH'].value_counts(), "\n\n",
      "Quality:\n\n", Wtotaal['quality'].value_counts(), "\n\n",
#      "total sulfur dioxide:\n\n", Wtotaal['total sulfur dioxide'].value_counts(), "\n\n",
#      "free sulfur dioxide:\n\n", Wtotaal['free sulfur dioxide'].value_counts(), "\n\n",
#      "residual sugar:\n\n", Wtotaal['residual sugar'].value_counts(), "\n\n",
#      "citric acid:\n\n", Wtotaal['citric acid'].value_counts(), "\n\n",
#      "chlorides:\n\n", Wtotaal['chlorides'].value_counts(), "\n\n",
#      "volatile acidity:\n\n", Wtotaal['volatile acidity'].value_counts(), "\n\n",
#      "fixed acidity:\n\n", Wtotaal['fixed acidity'].value_counts(), "\n\n",
      sep='')	
```

## Task Visualize dataset distribution by quality

```python
# show value distribution by quality
Wtotaal['quality'].describe()
```

```python
# Visualize dataset distribution by quality
fig = px.histogram(Wtotaal, x='quality')
fig.show()
```

## Task Visualize dataset distribution by features

```python
# Visualize dataset distribution from all features with plotply express
fig = px.scatter_matrix(Wtotaal, dimensions=["alcohol", "sulphates", "density", "pH", "quality"], color="quality")
fig.update_traces(diagonal_visible=False, selector=dict(type='scatter'))

fig.show()
```
